**Analysis 2: Correlation Analysis (Multivariate Analysis)**

 **Objective:**
To identify relationships between numerical variables.

 **Type of EDA:**
Non-Graphical, Multivariate Analysis

 **Method Used:**
Correlation Matrix (corr())

 **Observations:**

Some variables show strong positive correlation (close to +1)
Some show negative correlation (close to -1)
Weak correlations indicate little or no linear relationship

 **Conclusion:**
Highly correlated variables can be useful for prediction but may also cause multicollinearity, which should be handled carefully.

```python
print("Objective: To understand central tendency and spread of numerical features")
print("Type: Non-Graphical, Univariate Analysis\n")

# Summary statistics
summary = df.describe()
print(summary)

# Observations (Write in markdown in notebook)
print("\nObservations:")
print("- Compare mean and median (50%) to check skewness")
print("- Check standard deviation for variability")
print("- Look at min and max for outliers")

print("\nConclusion:")
print("- Helps identify distribution and possible outliers")
```

```python
print("Objective: To identify relationships between variables")
print("Type: Non-Graphical, Multivariate Analysis\n")

# Correlation matrix
correlation = df.corr(numeric_only=True)
print(correlation)

print("\nObservations:")
print("- Values close to +1 indicate strong positive correlation")
print("- Values close to -1 indicate strong negative correlation")
print("- Values near 0 indicate weak/no relationship")

print("\nConclusion:")
print("- Helps in feature selection and detecting multicollinearity")
```

**Analysis 3: Histogram (Univariate Analysis)**

 **Objective:**
To visualize the distribution of a numerical variable.

 **Type of EDA:**
Graphical, Univariate Analysis

** Method Used:**
Histogram plot

 **Observations:**

Data shows (normal / skewed) distribution
Right/left skew indicates imbalance
Peaks represent frequent values

 **Conclusion:**
The variable distribution helps in deciding whether normalization or transformation is needed.

```python
print("Objective: To visualize distribution of a numerical variable")
print("Type: Graphical, Univariate Analysis\n")

# Select a numerical column (change column name)
col = df.select_dtypes(include=np.number).columns[0]

plt.figure()
plt.hist(df[col], bins=20)
plt.title(f"Histogram of {col}")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.show()

print("\nObservations:")
print("- Check if distribution is normal or skewed")
print("- Identify peaks and spread of data")

print("\nConclusion:")
print("- Helps understand distribution and need for transformation")
```

**Analysis 4: Scatter Plot (Multivariate Analysis)**

 **Objective:**
To analyze the relationship between two numerical variables.

 **Type of EDA:**
Graphical, Multivariate Analysis

 **Method Used:**
Scatter Plot

 **Observations:**

Clear upward trend → positive relationship
Downward trend → negative relationship
Random pattern → no correlation

 **Conclusion:**
The relationship between variables helps in feature selection and model building.

```python
print("Objective: To analyze relationship between two variables")
print("Type: Graphical, Multivariate Analysis\n")

# Select two numerical columns
num_cols = df.select_dtypes(include=np.number).columns

if len(num_cols) >= 2:
    x = num_cols[0]
    y = num_cols[1]

    plt.figure()
    plt.scatter(df[x], df[y])
    plt.title(f"Scatter Plot of {x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()

    print("\nObservations:")
    print("- Upward trend → positive correlation")
    print("- Downward trend → negative correlation")
    print("- Random → no correlation")

    print("\nConclusion:")
    print("- Helps in identifying relationships for modeling")
else:
    print("Not enough numerical columns for scatter plot")
```
