Q1**What is 2 tailed and  tailed T Test ?**
Ans. A T-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups. The terms "1-tailed" and "2-tailed" refer to the directionality of the hypothesis being tested.

2-tailed T-test (Two-sided T-test)
Purpose: Used when you want to determine if there is any significant difference between the means of two groups, regardless of the direction of the difference. You are interested in whether one mean is simply different from the other.
Hypotheses:
Null Hypothesis (H₀): There is no significant difference between the two population means (e.g., μ₁ = μ₂).
Alternative Hypothesis (H₁): There is a significant difference between the two population means (e.g., μ₁ ≠ μ₂).
Critical Region: The critical region for rejecting the null hypothesis is split into two tails of the t-distribution. This means you are looking for extreme values on either side of the distribution.
When to use: When you don't have a prior expectation about the direction of the difference. For example, testing if a new drug changes blood pressure, without knowing if it will increase or decrease it.
1-tailed T-test (One-sided T-test)
Purpose: Used when you have a specific hypothesis about the direction of the difference between the means of two groups. You are interested in whether one mean is significantly greater than or significantly less than the other.
Hypotheses: There are two possible alternative hypotheses for a 1-tailed test:
Case 1 (Greater Than):
Null Hypothesis (H₀): The first population mean is less than or equal to the second (e.g., μ₁ ≤ μ₂).
Alternative Hypothesis (H₁): The first population mean is significantly greater than the second (e.g., μ₁ > μ₂).
Case 2 (Less Than):
Null Hypothesis (H₀): The first population mean is greater than or equal to the second (e.g., μ₁ ≥ μ₂).
Alternative Hypothesis (H₁): The first population mean is significantly less than the second (e.g., μ₁ < μ₂).
Critical Region: The critical region for rejecting the null hypothesis is entirely in one tail of the t-distribution (either the left or the right), depending on the direction of your alternative hypothesis.
When to use: When you have a strong theoretical reason or prior evidence to expect a difference in a particular direction. For example, testing if a new fertilizer increases crop yield.
Key Difference: The main difference lies in how the alternative hypothesis is formulated and how the critical region for statistical significance is distributed. A 2-tailed test is more conservative as it requires a stronger effect to achieve significance due to the alpha level being split across two tails. Generally, it's safer to use a 2-tailed test unless there's a clear, justified reason for a 1-tailed test.

**Q2.Explain what is Type I and Type II errors with an example.**
Ans. Certainly! Type I and Type II errors are fundamental concepts in hypothesis testing:

Type I Error (False Positive)
Definition: A Type I error occurs when the null hypothesis (H₀) is true, but you reject it. In simpler terms, you conclude there is a significant effect or difference when, in reality, there isn't one.
Symbol: Represented by the Greek letter alpha (α), which is also the significance level of the test.
Consequences: Can lead to unnecessary changes, resource allocation, or further research based on a non-existent effect.
Example:
Scenario: A pharmaceutical company develops a new drug for a common cold and conducts a clinical trial to see if it reduces the duration of the cold.
Null Hypothesis (H₀): The new drug has no effect on the duration of the common cold.
Alternative Hypothesis (H₁): The new drug does reduce the duration of the common cold.
Type I Error: The study concludes that the new drug significantly reduces cold duration (rejects H₀), when in fact, it has no real effect. This could lead the company to invest millions in a drug that doesn't work.
Type II Error (False Negative)
Definition: A Type II error occurs when the null hypothesis (H₀) is false, but you fail to reject it. In other words, there is a significant effect or difference, but your test fails to detect it.
Symbol: Represented by the Greek letter beta (β).
Consequences: Can lead to missed opportunities, failure to implement beneficial changes, or continued use of an inferior method/treatment.
Example:
Scenario: Continuing with the pharmaceutical company example.
Null Hypothesis (H₀): The new drug has no effect on the duration of the common cold.
Alternative Hypothesis (H₁): The new drug does reduce the duration of the common cold.
Type II Error: The study concludes that the new drug has no significant effect (fails to reject H₀), when in fact, it does effectively reduce cold duration. This could cause the company to abandon a genuinely effective drug, missing out on its potential benefits to patients and profitability.
Summary:
Type I Error: Claiming there's an effect when there isn't one (false alarm).
Type II Error: Missing an effect when there actually is one (missed opportunity).

**Q3What is power of statistical test?**
Ans. The power of a statistical test is the probability that the test will correctly reject a false null hypothesis. In simpler terms, it's the probability of finding an effect when there actually is an effect to be found. It is represented by 1 - β (where β is the probability of a Type II error).

Here's a breakdown:

Definition: The power of a test is its ability to detect a true effect (i.e., to correctly reject the null hypothesis when the alternative hypothesis is true).
Relationship to Type II Error: Power is inversely related to the probability of a Type II error (β). If a test has high power, it has a low probability of making a Type II error (a false negative).
High Power = Low β = Low chance of missing a true effect.
Low Power = High β = High chance of missing a true effect.
Importance:
Avoiding Missed Discoveries: A test with high power is more likely to detect genuine effects or differences, leading to more accurate scientific conclusions and preventing promising research avenues or treatments from being prematurely dismissed.
Ethical Considerations: In fields like medicine, a low-powered study might fail to show the effectiveness of a truly beneficial drug, potentially delaying its availability to patients.
Resource Efficiency: Conducting a study with insufficient power might be a waste of resources if it's unlikely to detect an effect even if one exists.
Factors Influencing Power: The power of a statistical test is influenced by several factors:
Effect Size: The magnitude of the difference or relationship you are trying to detect. A larger effect size is easier to detect, thus increasing power.
Sample Size: Larger sample sizes generally lead to higher power, as they provide more information and reduce sampling error.
Significance Level (α): A higher α (e.g., 0.10 instead of 0.05) increases the power but also increases the risk of a Type I error (false positive).
Variability (Standard Deviation): Less variability in the data (smaller standard deviation) increases power, as it makes true effects easier to discern from noise.
Type of Test: Some statistical tests are inherently more powerful than others for certain types of data or research questions.
In essence, power is about the sensitivity of your experiment – how good it is at picking up signals (true effects) amidst the noise (random variation).

**Q4.Explain input and out put characteristics for  stats.ttest_rel function in your own words**
Ans. The scipy.stats.ttest_rel function is used to calculate the T-test for related (paired) samples.

In simple terms, you'd use this function when you have two sets of measurements taken from the same subjects or matched pairs. For example, if you measure a group of people's blood pressure before a treatment and after the treatment, those are related samples.

Here's a breakdown of its input and output characteristics:

Inputs:
a (array_like): This is the first set of measurements. Think of it as the 'before' or 'first' group of observations. It should be a sequence of numbers (e.g., a list, NumPy array, or pandas Series).

b (array_like): This is the second set of measurements. This would be the 'after' or 'second' group of observations. Crucially, b must have the same shape and size as a, because each value in a corresponds directly to a value in b (they are paired).

axis (int or None, optional): This parameter specifies along which axis to compute the test if a and b are multi-dimensional arrays. If None (the default), it flattens the arrays and computes the test for all elements.

nan_policy ({'propagate', 'raise', 'omit'}, optional): This handles how NaN (Not a Number) values are treated:

'propagate' (default): If there are NaNs, they propagate to the output.
'raise': An error is raised if NaN is present.
'omit': NaN values are ignored. This means that if a pair has a NaN in either a or b, that entire pair is removed before computation.
alternative ({'two-sided', 'less', 'greater'}, optional): This specifies the alternative hypothesis. It determines whether you're performing a 2-tailed test or a 1-tailed test.

'two-sided' (default): Tests if the means are different (μ₁ ≠ μ₂).
'less': Tests if the mean of a is less than the mean of b (μ₁ < μ₂).
'greater': Tests if the mean of a is greater than the mean of b (μ₁ > μ₂).
Outputs:
The function returns an object (specifically, a scipy.stats._result_classes.TtestResult namedtuple) containing the following two main values:

statistic (float): This is the calculated t-statistic. It's a measure of the difference between the sample means relative to the variability within the samples. A larger absolute value of the t-statistic suggests a greater difference between the means.

pvalue (float): This is the two-tailed p-value (or one-tailed if alternative is set to 'less' or 'greater'). The p-value tells you the probability of observing a t-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true. A small p-value (typically less than 0.05) suggests that the observed difference is statistically significant, leading you to reject the null hypothesis.

**Q5.What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**
Ans.You're asking about a very important concept! While scipy.stats is excellent for performing statistical tests, it doesn't have a single, direct function named stats.power that universally calculates the power for all statistical tests. Power analysis is often handled by a more specialized library in Python, primarily statsmodels.

Within the statsmodels library, there's a submodule called statsmodels.stats.power that provides functions for power and sample size calculations for various statistical tests (e.g., t-tests, ANOVA, chi-squared tests, proportions).

Let's discuss the general characteristics of a power function and its output:

General Concept of Power Calculation Functions
Power functions typically take several arguments related to the test design and expected effect:

effect_size: This is the standardized difference or relationship you expect to find. For example, Cohen's d for t-tests, or f for ANOVA. A larger effect size generally means higher power.
alpha (α): The significance level (Type I error rate), commonly 0.05. A higher alpha increases power but also the chance of a Type I error.
nobs or n (Number of Observations/Sample Size): The sample size per group or total sample size. Larger sample sizes generally lead to higher power.
power: The desired statistical power (e.g., 0.80 or 80%).
ratio (for two-sample tests): The ratio of sample sizes between groups.
alternative: Specifies the type of alternative hypothesis (e.g., 'two-sided', 'less', 'greater'), similar to ttest_rel.
The key is that you typically provide three of these parameters and solve for the fourth. Common uses are:

Calculate Power: Given effect_size, alpha, and nobs, calculate power.
Calculate Sample Size: Given effect_size, alpha, and power, calculate nobs (the required sample size).
Output Range and Significance of Power
When you ask a power function to calculate the power of a test, the output (power) will be a float value typically ranging from 0 to 1.

Range:

0 (or close to 0): Implies a very low probability of detecting a true effect, even if one exists. The study is underpowered.
1 (or close to 1): Implies a very high probability of detecting a true effect. The study is very well-powered.
Commonly, a power of 0.80 (or 80%) is considered acceptable in many fields. This means there's an 80% chance of detecting a true effect of the specified size if it exists, and a 20% chance of making a Type II error (β = 0.20).
Significance:

High Power (e.g., ≥ 0.80): Indicates that your study design (given the effect_size, alpha, and nobs) has a good chance of correctly identifying a real effect if it's present. This is crucial for avoiding Type II errors (false negatives) and ensuring that your research is meaningful.
Low Power (e.g., < 0.50): Suggests that your study is unlikely to detect a true effect, even if it's substantial. Conducting such a study might be a waste of resources and time, as it has a high risk of producing a non-significant result simply due to insufficient power, rather than the absence of a real effect.
The significance of the power value lies in its role in experimental design and interpretation: It helps researchers design studies that are sensitive enough to detect effects of scientific or practical importance and avoid inconclusive results due to inadequate sample sizes.

To illustrate, if you were using statsmodels for a one-sample t-test power calculation, you might use TTestPower().solve_power.

from statsmodels.stats.power import TTestPower

# Example: Calculate power for a one-sample t-test
# with an effect size of 0.5, alpha of 0.05, and 30 observations.

power = TTestPower().solve_power(
    effect_size=0.5,
    nobs=30,
    alpha=0.05,
    alternative='two-sided'
)
print(f"Calculated Power: {power:.3f}")

# Example: Calculate required sample size for a power of 0.80
required_nobs = TTestPower().solve_power(
    effect_size=0.5,
    power=0.80,
    alpha=0.05,
    alternative='two-sided'
)
print(f"Required Sample Size: {int(required_nobs)}")
This shows how you would use a dedicated function from statsmodels for such calculations.
