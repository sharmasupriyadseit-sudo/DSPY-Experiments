**Attribute Selection & Filtering**

**Selected Attributes:**

Order_Date, Region, Category, Sales, Quantity, Profit

Removed Attributes:

Customer_Name → not required for analysis
Order_ID → unique identifier, not useful for trends

```python
df = df[['Order_Date', 'Region', 'Category', 'Sales', 'Quantity', 'Profit']]
```

**Reasoning:**

Focus is on sales trends and performance, not individual customers

**Data Cleaning**


```python
df.isnull().sum()
```

```python
df['Sales'].fillna(df['Sales'].mean(), inplace=True)
df['Profit'].fillna(df['Profit'].median(), inplace=True)
df.dropna(inplace=True)
```

**Reasoning:**

Mean/median preserves dataset distribution
Remaining null rows removed

**Remove Duplicates**

```python
df.drop_duplicates(inplace=True)
```

**Fix Data Types**

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

**Outlier Detection & Removal**

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Sales'] >= Q1 - 1.5*IQR) & (df['Sales'] <= Q3 + 1.5*IQR)]
```

**Reasoning:**

Removes extreme values that may distort analysis

**Data Wrangling**
**Feature Engineering**

Create new column:

```python
df['Sales_per_Unit'] = df['Sales'] / df['Quantity']
```

**Extract Date Features**

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

**Encoding Categorical Variables**

```python
df = pd.get_dummies(df, columns=['Region', 'Category'])
```

**Filtering Data**

Example:

```python
df = df[df['Sales'] > 0]
```

**Reasoning:**

Removes invalid transactions

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Sales', 'Profit']] = scaler.fit_transform(df[['Sales', 'Profit']])
```
