**Theory :**

**Matplotlib** is a multi-platform data visualization library built on NumPy arrays, and designed to work with the broader SciPy stack.

It was conceived by John Hunter in 2002, originally as a patch to IPython for enabling interactive MATLAB-style plotting via gnuplot from the IPython command line.

One of Matplotlib’s most important features is its ability to play well with many operating systems and graphics backends. Matplotlib supports dozens of backends and output types, which means you can count on it to work regardless of which operating system you are using or which output format you wish.

This cross-platform, everything-to-everyone approach has been one of the great strengths of Matplotlib. It has led to a large user base, which in turn has led to an active developer base and Matplotlib’s powerful tools and ubiquity within the scientific Python world.

In recent years, however, the interface and style of Matplotlib have begun to show their age. Newer tools like ggplot and ggvis in the R language, along with web visualization toolkits based on D3js and HTML5 canvas, often make Matplotlib feel clunky and old-fashioned.

**Create a collection of 100 random normal numbers with parameter mean 30, sd 4 store it in object summer_temp**



```python
import numpy as np

# Create a collection of 100 random normal numbers
mean = 30
sd = 4
num_samples = 100
summer_temp = np.random.normal(loc=mean, scale=sd, size=num_samples)

# Display the first few values to verify
print(summer_temp[:5])
```

**Sort above set and store in objrct sorted_summer_temp**

```python
sorted_summer_temp = np.sort(summer_temp)

# Display the first few values to verify
print(sorted_summer_temp[:5])
```

## Calculate Rolling Averages

### Subtask:
Calculate the average of every 25 consecutive elements from the `sorted_summer_temp` array. Store these rolling averages in a new object named `x`.


**Reasoning**:
The subtask requires calculating the average of every 25 consecutive elements from the `sorted_summer_temp` array and storing them in an object named `x`. The provided instructions detail the steps to achieve this, involving looping through the array, slicing 25-element windows, calculating the mean, and storing the results.



```python
rolling_averages = []
window_size = 25

for i in range(len(sorted_summer_temp) - window_size + 1):
    current_window = sorted_summer_temp[i : i + window_size]
    window_mean = np.mean(current_window)
    rolling_averages.append(window_mean)

x = np.array(rolling_averages)

# Display the first few values to verify
print(x[:5])
```

## Generate Line Plot

### Subtask:
Create a line plot using Matplotlib with the `x` object (containing the rolling averages) on the y-axis. The x-axis will represent the index or position of these averages. Ensure the plot has appropriate labels and a title.


**Reasoning**:
Generate a line plot using the calculated rolling averages stored in the 'x' object, applying appropriate labels and a title for clarity.



```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.plot(np.arange(len(x)), x)
plt.title('Rolling Average of Summer Temperatures (Window Size 25)')
plt.xlabel('Window Index')
plt.ylabel('Rolling Average Temperature')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate rolling averages for summer_temp and store in y
rolling_averages_y = []
window_size = 25 # Assuming the same window size as for x

for i in range(len(summer_temp) - window_size + 1):
    current_window = summer_temp[i : i + window_size]
    window_mean = np.mean(current_window)
    rolling_averages_y.append(window_mean)

y = np.array(rolling_averages_y)

# Display the first few values to verify
print("First 5 rolling averages for y:", y[:5])
```
